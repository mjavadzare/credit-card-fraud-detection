## log transformer

First we need to shift data so that there won't be any negative data anymore

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
X_train['V5'].min()
```

```python
def shift_to_right(X, eps=1): 
    return X + abs(X.min()) + eps
# epsilon for the exact min that will be zero, and then it'll be log0 
```

```python
shift_to_right(X_train['V5']).min()
```

```python
shifted = shift_to_right(X_train['V5'])
shifted = shifted.apply(np.log)
shifted.hist(bins=30)
```

```python
log_heavy_tail_pipeline = make_pipeline(
    FunctionTransformer(shift_to_right, feature_names_out='one-to-one', validate=False),
    FunctionTransformer(np.log, feature_names_out='one-to-one'),
    StandardScaler()
)
```

```python
fig, ax = plt.subplots()

ax.hist(X_train['Time'], bins=40, color='lightblue')
ax.set_xlabel('Time')
ax.set_ylabel('Count')
times = np.linspace(
    X_train['Time'].min(),
    X_train['Time'].max(),
    2000
).reshape(-1,1)

gamma1 = 1e-8  # 1 * 10^-8
gamma2 = 3e-9  # 3 * 10^-9
rbf_time1 = rbf_kernel(times, [[79000]], gamma=gamma1)
rbf_time2 = rbf_kernel(times, [[79000]], gamma=gamma2)

times = times.flatten()
rbf_time1 = rbf_time1.ravel()
rbf_time2 = rbf_time2.ravel()


ax2 = ax.twinx()
ax2.plot(times, rbf_time1, color='red', label='gamma= 1e-8')
ax2.plot(times, rbf_time2,  color='black', ls='--', label='gamma= 3e-9')

ax2.set_ylabel('RBF similarity to 79000')
ax2.legend(bbox_to_anchor=(1.45,1))

plt.title("Histogram of Time with RBF Similarity Curves")

plt.show()
```

## Column Transformer

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, FunctionTransformer
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.metrics.pairwise import rbf_kernel

Time_rbf_transformer = FunctionTransformer(
    rbf_kernel,
    feature_names_out='one-to-one',
    kw_args={
        'Y':[[79000]],
        'gamma':1e-8
    }
)
```

```python
from sklearn.impute import SimpleImputer

default_pipeline = make_pipeline(
    SimpleImputer(strategy='median'),
    StandardScaler(),
)
```

```python
from sklearn.preprocessing import QuantileTransformer

q_heavy_tail_pipeline = make_pipeline(
    QuantileTransformer(output_distribution='normal'),
    StandardScaler()
)
```

```python
from sklearn.decomposition import PCA

# V2, V5, V7 with Amount
vn_amount_pipeline = make_pipeline(
    StandardScaler(),
    PCA(n_components=2)
)
```

```python
all_transformation = ColumnTransformer(
    transformers=[
        ('Time_rbf', Time_rbf_transformer, ['Time']),
        ('heavy_tail', q_heavy_tail_pipeline, ['V1', 'V2', 'V3', 'V4', 'V5',
                                               'V6','V7', 'V8', 'V9', 'V10',
                                               'V11', 'V12', 'V14', 'V15',
                                               'V16', 'V17', 'V18', 'V19', 'V20',
                                               'V21', 'V22', 'V23', 'V24', 'V25',
                                               'V27', 'V28', 'Amount']),
        ('vn_amount', vn_amount_pipeline, ['V2', 'V5', 'V7'])
        
    ],
    remainder=default_pipeline,
)
```
